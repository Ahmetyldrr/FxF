```python
import os
import django
import pandas as pd
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'fxfutbol.settings')
django.setup()
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
```

```python
from analysis.models import MatchStats
fixture_data = MatchStats.objects.all().values()
df = pd.DataFrame(list(fixture_data))
df 
```

```python
df["stats"][0].keys()
```

```python
stats_df = pd.json_normalize(df['stats'])

stats_df
```

```python
columns = list(stats_df.columns)
columns
```

```python
match_columns = [col for col in columns if col.startswith("line")]
match_columns 
```

```python
mycols = ["lineups.match_id","lineups.response_code",
          "match_info.data.event.homeTeam.id","match_info.data.event.awayTeam.id",
          "match_info.data.event.homeTeam.name","match_info.data.event.awayTeam.name",
          "lineups.data.home.players","lineups.data.away.players",
          "lineups.data.home.missingPlayers","lineups.data.away.missingPlayers"]

data = stats_df[mycols]
data.columns = ["match_id","code_response",
                "home_id","away_id",
                "home_name","away_name",
                "home_players","away_players",
                "home_missing","away_missing"]
data
```

```python
datax = pd.json_normalize(data['home_players'])
datax["match_id"]=data["match_id"]
datax
```

```python
import numpy as np
match_ids = datax['match_id'].repeat(datax.iloc[:, :-1].apply(len, axis=)).reset_index(drop=True)

# TÃ¼m sÃ¼tunlardaki dict yapÄ±sÄ±nÄ± aÃ§mak ve uzun formata getirmek iÃ§in json_normalize kullanma
normalized_dfs = [pd.json_normalize(datax[col]) for col in datax.columns if col != 'match_id']

# TÃ¼m normalleÅŸtirilmiÅŸ DataFrame'leri birleÅŸtirme
result_df = pd.concat(normalized_dfs, ignore_index=True)

# `match_id` deÄŸerini normalleÅŸtirilmiÅŸ veriye ekleme
result_df['match_id'] = match_ids

# Sonucu gÃ¶sterme
result_df = result_df.fillna(0)
result_df = result_df[result_df["player.position"]!=0]
result_df
```
