```python
datam = []
for i in range(len(df_lite)):
    try:
        code = df_lite["stats"][i]["match_info"]["response_code"]
    except:
        code = ""
    try:
        idx = df_lite["stats"][i]['match_id']
    except:
        idx = ""
    try:
        lineup_code = df_lite["stats"][i]["lineups"]["response_code"]
    except:
        lineup_code = ""
    
    try:
        home = df_lite["stats"][i]["match_info"]["data"]["event"]["homeTeam"]["name"]
    except:
        home=""
    try:
        away = df_lite["stats"][i]["match_info"]["data"]["event"]["awayTeam"]["name"]
    except:
        away=""
        
    veri  = {
      'id':idx,
      'code':code,
        'lineup_code':lineup_code,
        'home':home,
        'away':away
        
    }
    
    datam.append(veri)
datam
```

```python
df = pd.DataFrame(datam)
df
```
